# Face Recognition Using facenet-pytorch

This project demonstrates how to perform face recognition using pre-trained models from the `facenet-pytorch` library. It allows recognizing and labeling faces in static images using pre-calculated embeddings.

---

## 📁 Project Structure

- `face_recognition.py` — main logic for detection, embedding, and visualization
- `embeddings.pt` — pre-saved face embeddings (database)
demonstration
