```python
# import libry

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import median_absolute_error
from sklearn.impute import SimpleImputer #clean
import numpy as np
```

```python
# Importing the dataset
dataset = pd.read_csv('MEN_SHOES.csv')
dataset.head(10)
```

```python
# check for missing values in variables

dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
dataset_enc = dataset.copy()
for col in dataset.columns:
    lb = LabelEncoder()
    dataset_enc[col] = lb.fit_transform(dataset[col].values)
dataset_enc
```

```python
#Concatenating the data back
data1 = pd.concat([ dataset_enc], axis = 1)
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=0, strategy='mean')
imp = imp.fit(data1)


modifieddata = imp.transform(data1)
print(modifieddata)
```
