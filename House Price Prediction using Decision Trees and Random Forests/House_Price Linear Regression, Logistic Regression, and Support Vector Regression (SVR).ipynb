```python
# import libry

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import median_absolute_error
from sklearn.impute import SimpleImputer #clean
import numpy as np
```

```python
# Importing the dataset
dataset = pd.read_csv('House_Price.csv')
dataset.head(10)
```

```python
# check for missing values in variables

dataset.isnull().sum()
```

```python
dataset.drop(['n_hos_beds'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
dataset_enc = dataset.copy()
for col in dataset.columns:
    lb = LabelEncoder()
    dataset_enc[col] = lb.fit_transform(dataset[col].values)
dataset_enc
```

```python
#Concatenating the data back
data1 = pd.concat([ dataset_enc], axis = 1)
```

```python
x=data1.iloc[:,:1]#  Ø¬Ù…ÙŠØ¹ Ø§Ù„Ø¯Ø§ØªØ§ Ù…Ø§ Ø¹Ø¯Ø§ Ø§Ø®Ø± Ø¹Ù…ÙˆØ¯xØ§ÙŠ Ø§Ù†
y=data1.iloc[:,1]#   Ø§Ø®Ø± Ø¹Ù…ÙˆØ¯yØ§ÙŠ Ø§Ù†
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x)
print(scaler.mean_)
newx = scaler.transform(x)
print(newx)
```

```python
#Splitting data

x_train, x_test, y_train, y_test = train_test_split(newx, y, test_size=0.2, random_state=44, shuffle =True)

#----------------------------------------------------
```

```python
#Applying Linear Regression Model

LinearRegressionModel = LinearRegression(fit_intercept=True, normalize=True,copy_X=True,n_jobs=-1)
LinearRegressionModel.fit(x_train, y_train)
```

```python
#Calculating Details
print('Linear Regression Train Score is : ' , LinearRegressionModel.score(x_train, y_train))
print('Linear Regression Test Score is : ' , LinearRegressionModel.score(x_test, y_test))
print('Linear Regression Coef is : ' , LinearRegressionModel.coef_)
print('Linear Regression intercept is : ' , LinearRegressionModel.intercept_)
```

```python
#Calculating Prediction
y_pred = LinearRegressionModel.predict(x_test)
print('Predicted Value for Linear Regression is : ' , y_pred[:10])
```

```python
#Calculating Mean Absolute Error
MAEValue = mean_absolute_error(y_test, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Absolute Error Value is : ', MAEValue)

#----------------------------------------------------
#Calculating Mean Squared Error
MSEValue = mean_squared_error(y_test, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Squared Error Value is : ', MSEValue)

#----------------------------------------------------
#Calculating Median Squared Error
MdSEValue = median_absolute_error(y_test, y_pred)
print('Median Squared Error Value is : ', MdSEValue )
```

# 2.	Logistic Regression

```python
import pandas as pd

dataset = pd.read_csv('House_Price.csv')

dataset.head(20)
```

```python
# check for missing values in variables

dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
dataset_enc = dataset.copy()
for col in dataset.columns:
    lb = LabelEncoder()
    dataset_enc[col] = lb.fit_transform(dataset[col].values)
dataset_enc
```

```python
#Concatenating the data back
dataset = pd.concat([ dataset_enc], axis = 1)
```

```python
X= dataset.drop(['price'], axis=1)

y = dataset['price']
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
print(scaler.mean_)
newX = scaler.transform(X)
print(newX)
```

```python
#Splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(newX, y, test_size=0.2, random_state=44, shuffle =True)
```

```python
#Applying Logistic  Regression Model

from sklearn.linear_model import LogisticRegression

clss = LogisticRegression(random_state=10, solver='lbfgs' , max_iter= 1000 , C = 0.5 , tol = 0.01)
#clss = LogisticRegression(random_state=10, solver='liblinear')
#clss = LogisticRegression(random_state=2, solver='saga')

clss.fit(X_train, y_train)

# Predicting the Test set results
y_pred = clss.predict(X_test)
y_pred
clss.n_iter_
clss.classes_


```

```python
#probability of all values
pr = clss.predict_proba(X_test)[0:10,:]
pr
```

```python
#probability of zeros
pr = clss.predict_proba(X_test)[0:10,0]
pr
```

```python
#probability of ones
pr = clss.predict_proba(X_test)[0:10,1]
pr
```

```python
#  Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm

```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)

```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
from sklearn.metrics import median_absolute_error
median_absolute_error(y_test, y_pred)
```
