```python
import numpy as np
import pandas as pd

dataset = pd.read_csv('Earthquakes.csv')

dataset.head(20)
```

```python
# check for missing values in variables

dataset.isnull().sum()
```

```python
#داتا string التعامل مع
#Splitting the data
a = dataset.iloc[:, [0]]  # تحتوي علي هذه الارقامaاي ان
b = dataset.iloc[:,[ 1,2,3,4]] # تحتوي علي هذه الارقامbاي ان
a
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
a_enc = a.copy()
for col in a.columns:
    lb = LabelEncoder()
    a_enc[col] = lb.fit_transform(a[col].values)
a_enc
```

```python
#Concatenating the data back
data1 = pd.concat([a_enc, b], axis = 1)
```

```python
# check for missing values in variables

data1.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp = imp.fit(data1)
dataset = imp.transform(data1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
dataset = sc.fit_transform(dataset)


X = dataset[:, :-2]
y = dataset[:, -2]
```

```python
#----------------------------------------------------
#Splitting data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle =True)
```

```python
#----------------------------------------------------
#Applying SVR Model
from sklearn.svm import SVR

'''
sklearn.svm.SVR(kernel=’rbf’, degree=3, gamma=’auto_deprecated’, coef0=0.0, tol=0.001,
                C=1.0, epsilon=0.1, shrinking=True, cache_size=200, verbose=False,max_iter=-1)
'''

SVRModel = SVR(C = 20,epsilon=0.01,kernel = 'rbf') # it  can be : linear, poly, rbf, sigmoid, precomputed
SVRModel.fit(X_train, y_train)

#Calculating Details
print('SVRModel Train Score is : ' , SVRModel.score(X_train, y_train))
print('SVRModel Test Score is : ' , SVRModel.score(X_test, y_test))
print('----------------------------------------------------')
```
